```python
from faker import Faker
import pandas as pd
import random
from datetime import datetime
import openpyxl
```

```python
fake = Faker('pt_BR')
Faker.seed(4400)
```

```python
df_folders = "./datasets/"
```

# Clientes

```python
size = 10
dados = {
    'ID': [ _ for _ in range(size)],
    'NAME': [fake.name() for _ in range(size)],
    'BIRTHDAY': [fake.date_of_birth(minimum_age=12, maximum_age=90) for _ in range(size)],
    'PHONE': [fake.phone_number() for _ in range(size)],
    'EMAIL': [fake.email() for _ in range(size)],
    'CITY': [fake.city() for _ in range(size)],
    'UF': [fake.estado_sigla() for _ in range(size)],
    'ADDRESS': [fake.street_address() for _ in range(size)],
    'NR_DOC': [fake.cpf() for _ in range(size)],
    'JOB_TITLE': [fake.job() for _ in range(size)],
    'INSERT_DATE': [fake.date_time_this_decade() for _ in range(size)],
    'LAST_UPDATE': [fake.date_time_this_decade() for _ in range(size)]
}
consumers_df = pd.DataFrame(dados)
consumers_df.head()
```

```python
consumers_df.to_csv(df_folders + 'consumers.csv', index=False, sep=',', header=True)
```

# Produtos

```python
products_df = pd.read_csv('my-products.csv')
```

```python
size = 100
dados = {
  'INSERT_DATE': [fake.date_time_this_decade() for _ in range(size)],
  'LAST_UPDATE': [fake.date_time_this_decade() for _ in range(size)]
}

products_df = products_df.assign(**dados)
products_df.head()
```

```python
products_df.to_csv('products.csv', index=False, sep=',', header=True)
```

# Lojas

```python
size = 20
dados = {
    'ID': [ _ for _ in range(size)],
    'NAME': [fake.company() for _ in range(size)],
    'PHONE': [fake.phone_number() for _ in range(size)],
    'CITY': [fake.city() for _ in range(size)],
    'UF': [fake.estado_sigla() for _ in range(size)],
    'ADDRESS': [fake.street_address() for _ in range(size)],
    'NR_DOC': [fake.cnpj() for _ in range(size)],
    'INSERT_DATE': [fake.date_time_this_decade() for _ in range(size)],
    'LAST_UPDATE': [fake.date_time_this_decade() for _ in range(size)]
}

stores_df = pd.DataFrame(dados)
stores_df.head()
```

```python
stores_df.to_csv('stores.csv', index=False, sep=',', header=True)
```

# Vendas

```python
size = 100
dados = {
  'SALE_ID': [ _ for _ in range(size)],
  'PRODUCT_ID': [random.choice(products_df['ID']) for _ in range(size)],
  'AMOUNT': [random.randrange(1, 10) for _ in range(size)],
  'DISCOUNTED': 0,
  'STORE_ID': [random.choice(stores_df['ID']) for _ in range(size)],
  'CONSUMER_ID': [random.choice(consumers_df['ID']) for _ in range(size)],
  'SALE_DATE': [fake.date_time_this_decade() for _ in range(size)]
}
sales_df = pd.DataFrame(dados)
sales_df.head()
```
